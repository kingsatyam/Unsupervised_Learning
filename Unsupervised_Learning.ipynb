```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X1= np.array([7,3,5,7,8,2,4,1,9,5])
X2= np.array([22,36,11,8,90,12,33,57,82,17])
data = list(zip(X1,X2))
print(data)
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 2)
model.fit(data)
plt.scatter(X1,X2,c=model.labels_)
plt.title("K-Means")
plt.show()
```

```python
import pandas as pd 
df = pd.read_csv("iris.csv")
y1=df["petal.length"]
y2=df["petal.width"]
```

```python
data = list(zip(y1,y2))
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 22)
model.fit(data)
plt.scatter(y1,y2,c=model.labels_)
plt.title("K-Means")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters = 2)
model.fit(data)
plt.scatter(y1,y2,c=model.labels_)
plt.title("AgglomerativeClustering")
plt.show()
```

```python
import plotly.express as px
fig = px.scatter_3d(df,x="sepal.length",y="sepal.width",z="petal.length",color=model.labels_)
fig
```
